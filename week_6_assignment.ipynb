Q1. Explain the roles of the Driver, Cluster Manager, and Executor in a Spark application.

Answer:

Driver: Controls the Spark application, creates the SparkSession, schedules tasks, and collects results.
Cluster Manager: Allocates resources (CPU and memory) for Spark applications. Examples include Standalone, YARN, Mesos, and Kubernetes.
Executor: Executes tasks assigned by the Driver, stores intermediate data, and returns results.

Q2. How does Spark's Lazy Evaluation strategy improve performance when chain-processing large datasets?

Answer:

Spark records all transformations but does not execute them immediately. Execution starts only when an action is called. This allows Spark to optimize the execution plan using a Directed Acyclic Graph (DAG), reducing unnecessary computations and improving performance.

Q7. Explain Lineage Graph (DAG).

Answer:

Spark maintains a Lineage Graph (DAG) that records all transformations. If a worker node fails, Spark recomputes only the lost partitions instead of the entire dataset, providing fault tolerance.

Q9. Explain Predicate Pushdown.

Answer:

Predicate Pushdown is an optimization where Spark sends filter conditions directly to the Parquet reader. Only matching rows are read from disk, reducing memory usage and improving performance.

Q11. What is the difference between Transformations and Actions? Give two examples of each.

Answer:

In Apache Spark, Transformations and Actions are two types of operations performed on DataFrames or RDDs.

Transformations create a new DataFrame or RDD from an existing one without executing the operation immediately. Spark records these operations and executes them only when an action is called (Lazy Evaluation).
Actions trigger the execution of all pending transformations and either return a result to the Driver or write data to storage.

Q15. Why use .show(5) instead of .collect()?

Answer:

.show(5) displays only the first five rows, making it memory-efficient. .collect() loads the entire dataset into the Driver's memory, which can cause memory overflow or application failure for very large datasets.
